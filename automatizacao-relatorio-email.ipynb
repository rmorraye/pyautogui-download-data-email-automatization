```python
import pandas as pd
from datetime import date as dt
```

```python
tabela = pd.read_excel(r"Vendas - Dez.xlsx")

display(tabela)
```

```python
quantidade = tabela["Quantidade"].sum()

faturamento = tabela["Valor Final"].sum()

print(f"qntdd: {quantidade} / faturamento: {faturamento} ")
```

```python
# juntando cada loja atraves do somatorio de outros valores
df_merging_lojas = tabela.groupby(tabela["ID Loja"]).sum()
display(df_merging_lojas)


### LOJA com MAIOR FATURAMENTO ###
# pegando o *index da linha* da loja com maior faturamento 
index_max = df_merging_lojas["Valor Final"].idxmax()

display(index_max) # printando o index (a loja nesse caso)
display(df_merging_lojas.loc[index_max]["Valor Final"]) # maior faturamento da loja

maior_fat_loja = index_max
maior_fat_valor = df_merging_lojas.loc[index_max]["Valor Final"]
maior_fat_qntd = df_merging_lojas.loc[index_max]["Quantidade"]


### LOJA com MENOR FATURAMENTO ###
index_min = df_merging_lojas["Valor Final"].idxmin()

display(index_min) # printando o index (a loja nesse caso)
display(df_merging_lojas.loc[index_min]["Valor Final"]) # valor menor faturamento da loja 

menor_fat_loja = index_min
menor_fat_valor = df_merging_lojas.loc[index_min]["Valor Final"]
menor_fat_qntd = df_merging_lojas.loc[index_min]["Quantidade"]

### MÉDIA de FATURAMENTO das LOJAS ###
med_fat = df_merging_lojas["Valor Final"].mean()
med_qntd = df_merging_lojas["Valor Final"].mean()

```

```python
# enviar um email

pg.hotkey("ctrl", "t")
pc.copy("https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox")
pg.hotkey("ctrl", "v")
pg.press("Enter")

time.sleep(10)

# clicar no botao +
pg.click(x=85, y=209)

time.sleep(5)

# escrever o destinatario
pg.write("destinatario@gmail.com")
pg.press("tab")

# clicar no campo assunto e escrever 
pg.click(x=274, y=426)
pg.write("Relatorio - Vendas das lojas")

# clicar no campo da mensagem e escrever a mensagem
pg.click(x=317, y=467)

time.sleep(2)

# montagem do texto para o email

data = dt.today()
data_formatada = data.strftime("%d/%m/%Y")

text = f"""Bom dia, Superior

 *Relatório ({data_formatada})*

Quantidade total de itens vendidos: {quantidade:,} itens.
Faturamento total: R$ {faturamento:,.2f}.

Média da quantidade total de itens vendidos das lojas: {int(med_qntd):,} itens.
Média de faturamento das lojas: R$ {med_fat:,.2f}.

Maior faturamento (loja): {maior_fat_loja}.
Quantidade total de itens vendidos da loja {maior_fat_loja}: {maior_fat_qntd:,} itens.
Faturamento total da loja {maior_fat_loja}: R$ {maior_fat_valor:,.2f}.

Menor faturamento (loja): {menor_fat_loja}.
Quantidade total de itens vendidos da loja {menor_fat_loja}: {menor_fat_qntd} itens.
Faturamento total da loja {menor_fat_loja}: R$ {menor_fat_valor:,.2f}.

Att. Seu Funcionario """

pc.copy(text)
pg.hotkey("ctrl", "v")

### CLICAR EM ENVIAR E-MAIL ###
pg.click(x=200, y=698)
```
